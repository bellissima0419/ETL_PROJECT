```python
MY_API_KEY = APIKey
HEADERS = {'Authorization': 'bearer %s' % MY_API_KEY}
```

```python
# us_cannabis_url = "https://api.yelp.com/v3/businesses/search"
base_url = "https://api.yelp.com/v3/businesses/search"
```

```python
atl_PARAMETERS = {'location': 'Atlanta ga united states',
                  'limit': 50,
                  'radius': 40000
                 }
```

```python
atl_response = requests.get(url = base_url,
                        params = atl_PARAMETERS,
                        headers = HEADERS)
```

```python
for i in range(len(bins)):
    temp_urls.append(response_url + "&offset=" + f"{str(bins[i])}")
```

```python
print(temp_urls[0],'\n', temp_urls[-1])
```

```python
atl_biz_df = pd.DataFrame()
```

```python
atl_biz_df['id'] = ''
atl_biz_df['alias'] = ''
atl_biz_df['name'] = ''
atl_biz_df['categories'] = ''
atl_biz_df['review_count'] = ''
atl_biz_df['rating'] = ''
atl_biz_df['lat'] = ''
atl_biz_df['long'] = ''
# us_cannabis['transactions'] = ''
atl_biz_df['address'] = ''
atl_biz_df['city'] = ''
atl_biz_df['zip_code'] = ''
atl_biz_df['country'] = ''
atl_biz_df['state'] = ''
atl_biz_df['phone'] = ''
```

```python
# index_ = 0
for i in range(len(temp_urls)):
    print(temp_urls[i])
```

```python
index_ = 0
```

```python
i = 0
j = 0

try:
    for i in range(len(temp_urls)):
        temp_response = requests.get(url = temp_urls[i], headers = HEADERS)
        print(f"temp_response: {temp_response}")
        
        temp_response_json = temp_response.json()
        
        print("=======================================================")
        print(f"OUTTER LOOP BEGGINS: URL: {temp_response.url}" )

        for j in range(len(temp_response_json['businesses'])):
            print("=======================================================")
            print(f"inner loop (building df): biz index: {index_}: {temp_response_json['businesses'][j]['name']}")
              
            atl_biz_df.set_value(index_, "id", temp_response_json['businesses'][j]['id'])
            atl_biz_df.set_value(index_, "alias", temp_response_json['businesses'][j]['alias'])

            atl_biz_df.set_value(index_, "name", temp_response_json['businesses'][j]['name'])
            atl_biz_df.set_value(index_, "categories", ', '.join([val['title'] for val in temp_response_json['businesses'][j]['categories']]) )        

            atl_biz_df.set_value(index_, "review_count", temp_response_json['businesses'][j]['review_count'])
            atl_biz_df.set_value(index_, "rating", temp_response_json['businesses'][j]['rating'])

            atl_biz_df.set_value(index_, "lat", temp_response_json['businesses'][j]['coordinates']['latitude'])
            atl_biz_df.set_value(index_, "long", temp_response_json['businesses'][j]['coordinates']['longitude'])

            atl_biz_df.set_value(index_, "address", ' '.join(temp_response_json['businesses'][j]['location']['display_address']))
            atl_biz_df.set_value(index_, "city", temp_response_json['businesses'][j]['location']['city'])
                  
            atl_biz_df.set_value(index_, "zip_code", temp_response_json['businesses'][j]['location']['zip_code'])
            atl_biz_df.set_value(index_, "country", temp_response_json['businesses'][j]['location']['country'])
                  
            atl_biz_df.set_value(index_, "state", temp_response_json['businesses'][j]['location']['state'])
            atl_biz_df.set_value(index_, "phone", temp_response_json['businesses'][j]['phone'])

            index_ += 1
              
        time.sleep(3)

except:
    print("Something went WRONG")

print(f"Database ready: records entered:{index_}")
```
